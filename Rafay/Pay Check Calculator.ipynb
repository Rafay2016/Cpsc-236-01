```python
Hours = float(input("Hours Worked:"))
HourlyPayRate = float(input("Hourly Pay Rate:"))
GrossPay= Hours * HourlyPayRate
TaxRate = 18
TaxAmount = GrossPay * (TaxRate / 100)
TakeHome = GrossPay - TaxAmount

print("\nPay Check Calculator")
print("Hours Worked:", Hours)
print("Hourly Pay Rate:", HourlyPayRate)
print("Gross Pay:", round(GrossPay, 2))
print("Tax Rate:", str(TaxRate) + "%")
print("Tax Amount:", round(TaxAmount, 2))
print("Take Home Pay:", round(TakeHome, 2))
```
