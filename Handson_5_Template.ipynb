## 7.1 Pig Dice Rules
Create a program that reads a list of rules from a file and displays them.

### Console:
```powershell
Pig Dice Rules:
*	See how many turns it takes you to get to 20.
*	Turn ends when player rolls a 1 or chooses to hold.
*	If you roll a 1, you lose all points earned during the turn.
*	If you hold, you save all points earned during the turn.
```

### Specifications:
- Use the attached text file named `pig_dice_rules.txt` in the corresponding Directory.
- Your program should read the text file and display it on the console.


````python
import java.io.BufferedReader;
import java.io.FileReader;

public class PigDiceRules {

    public static void main(String[] args) {

        System.out.println("Pig Dice Rules:");

        try {
            BufferedReader br = new BufferedReader(
                    new FileReader("pig_dice_rules.txt"));

            String line;

            while ((line = br.readLine()) != null) {
                System.out.println("*\t" + line);
            }

            br.close();

        } catch (Exception e) {
            System.out.println("Error reading file.");
        }
    }
}
````

## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path 
show - Show all items 
drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. 
Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. 
Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit 
Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import java.io.*;
import java.util.*;

public class WizardInventory {

    static final String ALL_ITEMS_FILE = "wizard_all_items.txt";
    static final String INVENTORY_FILE = "wizard_inventory.txt";

    static ArrayList<String> inventory = new ArrayList<>();
    static final int MAX_ITEMS = 4;

    public static void main(String[] args) {
        loadInventory();
        Scanner sc = new Scanner(System.in);
        String command = "";

        System.out.println("The Wizard Inventory Program");

        while (!command.equals("exit")) {

            System.out.println("\nCOMMAND MENU");
            System.out.println("walk - Walk down the path");
            System.out.println("show - Show all items");
            System.out.println("drop - Drop an item");
            System.out.println("exit - Exit program");

            System.out.print("\nCommand: ");
            command = sc.nextLine();

            if (command.equals("walk"))
                walk(sc);
            else if (command.equals("show"))
                showItems();
            else if (command.equals("drop"))
                dropItem(sc);
        }

        System.out.println("Bye!");
    }

    // load inventory from file
    public static void loadInventory() {
        try {
            BufferedReader br = new BufferedReader(new FileReader(INVENTORY_FILE));
            String line;

            while ((line = br.readLine()) != null) {
                inventory.add(line);
            }

            br.close();
        } catch (Exception e) {
            // ignore if file doesn't exist yet
        }
    }

    // save inventory to file
    public static void saveInventory() {
        try {
            PrintWriter pw = new PrintWriter(new FileWriter(INVENTORY_FILE));

            for (String item : inventory) {
                pw.println(item);
            }

            pw.close();
        } catch (Exception e) {
            System.out.println("Error saving file.");
        }
    }

    // walk command
    public static void walk(Scanner sc) {
        ArrayList<String> allItems = new ArrayList<>();

        try {
            BufferedReader br = new BufferedReader(new FileReader(ALL_ITEMS_FILE));
            String line;

            while ((line = br.readLine()) != null) {
                allItems.add(line);
            }

            br.close();
        } catch (Exception e) {
            System.out.println("Error reading items.");
            return;
        }

        Random rand = new Random();
        String item = allItems.get(rand.nextInt(allItems.size()));

        System.out.println("While walking down a path, you see " + item + ".");
        System.out.print("Do you want to grab it? (y/n): ");
        String choice = sc.nextLine();

        if (choice.equalsIgnoreCase("y")) {
            if (inventory.size() >= MAX_ITEMS) {
                System.out.println("You can't carry any more items. Drop something first.");
            } else {
                inventory.add(item);
                saveInventory();
                System.out.println("You picked up " + item + ".");
            }
        }
    }

    // show command
    public static void showItems() {
        int i = 1;
        for (String item : inventory) {
            System.out.println(i + ".\t" + item);
            i++;
        }
    }

    // drop command
    public static void dropItem(Scanner sc) {
        System.out.print("Number: ");

        try {
            int num = Integer.parseInt(sc.nextLine());

            if (num < 1 || num > inventory.size()) {
                System.out.println("Invalid number.");
                return;
            }

            String removed = inventory.remove(num - 1);
            saveInventory();
            System.out.println("You dropped " + removed + ".");

        } catch (Exception e) {
            System.out.println("Invalid number.");
        }
    }
}
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:
```powershell

COMMAND MENU

Monthly average:	5550.5

```

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import java.io.*;
import java.util.*;

public class MonthlySales {

    static String fileName = "monthly_sales.csv";
    static LinkedHashMap<String, Double> sales = new LinkedHashMap<>();

    public static void main(String[] args) {
        loadFile();

        Scanner sc = new Scanner(System.in);
        String command = "";

        System.out.println("Monthly Sales program");

        while (!command.equals("exit")) {
            System.out.println("\nCOMMAND MENU");
            System.out.println("monthly - View monthly sales");
            System.out.println("yearly  - View yearly sumary");
            System.out.println("edit    - Edit sales for a month");
            System.out.println("exit    - Exit program");

            System.out.print("\nCommand: ");
            command = sc.nextLine();

            if (command.equals("monthly"))
                displayMonthly();
            else if (command.equals("yearly"))
                displayYearly();
            else if (command.equals("edit"))
                editSales(sc);
        }

        System.out.println("Bye!");
    }

    public static void loadFile() {
        try {
            BufferedReader br = new BufferedReader(new FileReader(fileName));
            String line;

            while ((line = br.readLine()) != null) {
                String[] parts = line.split(",");
                sales.put(parts[0], Double.parseDouble(parts[1]));
            }

            br.close();
        } catch (Exception e) {
            System.out.println("File error.");
        }
    }

    public static void displayMonthly() {
        for (String month : sales.keySet()) {
            System.out.println(month + " - " + sales.get(month).intValue());
        }
    }

    public static void displayYearly() {
        double total = 0;

        for (double value : sales.values()) {
            total += value;
        }

        double average = total / sales.size();

        System.out.println("Yearly total:\t" + (int) total);
        System.out.printf("Monthly average:\t%.2f\n", average);
    }

    public static void editSales(Scanner sc) {
        System.out.print("Three-letter Month: ");
        String month = sc.nextLine();

        if (!sales.containsKey(month)) {
            System.out.println("Invalid month.");
            return;
        }

        System.out.print("Sales Amount: ");
        double amount = Double.parseDouble(sc.nextLine());

        sales.put(month, amount);
        saveFile();

        System.out.println("Sales amount for " + month + " was modified.");
    }

    public static void saveFile() {
        try {
            PrintWriter pw = new PrintWriter(new FileWriter(fileName));

            for (String month : sales.keySet()) {
                pw.println(month + "," + sales.get(month));
            }

            pw.close();
        } catch (Exception e) {
            System.out.println("Save error.");
        }
    }
}
````
